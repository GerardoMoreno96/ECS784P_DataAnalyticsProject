```python
import pandas as pd
```

```python
df_sales = pd.read_csv("m5/bayesian_dataset.csv")
```

```python
df_sales[["Food_Sales"]].info
```

```python
df_sales.head()
```

```python
t_df=df_sales
```

```python
t_df=t_df.replace(to_replace ="d_", 
                 value ="",regex=True) 
```

```python
t_df["Day"] =pd.to_numeric(t_df.Day, errors='coerce')
```

```python
t_df.dtypes
```

```python
t_df
```

```python
t_df['Foods_items_Sold'] = pd.cut(x=t_df['Food_Sales'], bins=[0,8000,16000,24000,32000,40000], labels=['0Kto8K','8Kto16K', '16Kto24K','24Kto32K','32Kto40K'])
t_df['Hobbies_Items_Sold'] = pd.cut(x=t_df['Hobbies_Sales'], bins=[0,1000, 2000, 3000, 4000,5000,6000],labels=['0Kto1K','1Kto2K', '2Kto3K','3Kto4K','4Kto5k','5Kto6K'] )

t_df['Household_items_Sold'] = pd.cut(x=t_df['Household_Sales'], bins=[0,3000, 6000, 9000, 12000,15000], labels=['0Kto3K','3Kto6K', '6Kto9K','9Kto12K','12Kto15K'])
t_df['Total_Revenue_in_Cash'] = pd.cut(x=t_df['Revenue'], bins=[0,30500, 61000, 91500, 122000,152500,183000], labels=['0Kto30.5K','30.5Kto61K','61Kto91.5K', '91.5Kto122K','122Kto152.5K','152.5Kto183K'])
t_df["Day"]= pd.cut(x=t_df['Day'], bins=[0,365, 730, 1095, 1460,1825,2190], labels=['Year1','Year2','Year3', 'Year4','Year5','Year6'])

#t_df['Revenue'] = pd.cut(x=t_df['Revenue'], bins=[0,30000, 60000, 90000, 120000,150000,180000], labels=['0Kto30.5K','30.5Kto61K','61Kto91.5K', '91.5Kto122K','122Kto152.5K','152.5Kto183K'])

```
